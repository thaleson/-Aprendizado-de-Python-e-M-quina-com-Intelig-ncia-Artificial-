# **BASE DADOS MERCADO**

```python
import pandas as pd 
from apyori import  apriori
```

```python
base_mercado=pd.read_csv('C:/Users/thale/Python/Algoritmo de associaçao/mercado.csv' , header=None )
base_mercado
```

```python
transacoes=[]
for i in range(len(base_mercado)):
    print(base_mercado.values[i,0 ])

    transacoes.append([str(base_mercado.values[i,j])for  j in range(base_mercado.shape[1])])
```

```python
type(transacoes )   
```

```python
regras=apriori(transacoes , min_support=0.3,min_confidence=0.8,min_lift=2)
resultados=list(regras )
```

```python
A=[]
B=[]

suporte=[]
confiança=[]
lift=[]
for resultado in resultados:
    # print(resultado)
    s=resultado[1]
    # print(s)
     
    result=resultado[2]
    # print(result)
    for results in result:
       print(results)
       a=list(results[0])
       b=list(results[1])
       c=results[2]
       l=results[3]
       print(a,'---',b ,'---',c,'---',l)
       A.append(a)
       B.append(b)
       suporte.append(s)
       confiança.append(c)
       lift.append( l)
```

```python
regras=apriori(transacoes, min_support=0.003, min_confidence=0.2,min_lift=3)
resultados=list(regras)
len(resultados)
```

```python
A = []
B = []

suporte = []
confiança = []
lift = []
for resultado in resultados:
    # print(resultado)
    s = resultado[1]
    # print(s)

    result = resultado[2]
    # print(result)
    for results in result:
       print(results)
       a = list(results[0])
       b = list(results[1])
       c = results[2]
       l = results[3]
       print(a, '---', b, '---', c, '---', l)
       A.append(a)
       B.append(b)
       suporte.append(s)
       confiança.append(c)
       lift.append(l)
```

```python
len_A = len(A)
len_B = len(B)
len_suporte = len(suporte)
len_confianca = len(confiança)
len_lift = len(lift)

# Verifique os comprimentos
print(len_A, len_B, len_suporte, len_confianca, len_lift)
```

```python
rules_df = pd.DataFrame({'A':A,'B':B,'suporte':suporte,'confiança':confiança,'lift':lift})
rules_df.sort_values(by='confiança', ascending=False)
```
